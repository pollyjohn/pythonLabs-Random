```python
import csv
from itertools import cycle


youth_list = young_women + young_man
young_adults_repeated = young_adults * 2 


rows = []


youth_iterator = cycle(youth_list)
young_adults_iterator = cycle(young_adults_repeated)
elders_quorum_iterator = cycle(elders_quorum)
relief_society_iterator = cycle(relief_society)

while relief_society:
    youth = next(youth_iterator)
    young_adult = next(young_adults_iterator)
    elder_quorum = next(elders_quorum_iterator)
    relief_society_member = next(relief_society_iterator)

    row = [youth, young_adult, elder_quorum, relief_society_member]
    rows.append(row)
    if relief_society_member == relief_society[-1]: 
        break

csv_file = 'output.csv'

with open(csv_file, 'w', newline='') as file:
    writer = csv.writer(file)
    
    writer.writerow(['Youth', 'Young Adult', 'Elder Quorum', 'Relief Society'])
    
    writer.writerows(rows)

print(f"CSV file '{csv_file}' has been created.")
```
